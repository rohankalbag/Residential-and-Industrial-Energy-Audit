## Functions and Main Code

```python
import matplotlib.pyplot as plt,numpy as np

#Changing plot size
plt.rcParams['figure.figsize'] = [18, 10]
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 20
timeline = np.arange(0,24,0.1) # 24 hours divided into intervals of 6 minutes

def add(x,y):
    #to add two lists x and y
    a = list(x)
    for i in range(len(a)):
        a[i] = x[i] + y[i]
    return a
    
def digital(n,arr):
    #to create constant power consumption arrays for each device under constant power assumption
    darr = list(arr)
    if(n==1):
        for i in range(0,len(darr)):
            if(8 <= darr[i] <= 10) or (12 <= darr[i]  <= 14) or (16 <= darr[i]  <= 18) or (20<= darr[i] <=22):
                darr[i] = 100.
            else:
                darr[i] = 0.
        return darr
    if(n==2):
        for i in range(0,len(darr)):
            if(8 <= darr[i] <= 10) or (12 <= darr[i]  <= 14) or (16 <= darr[i]  <= 18) or (20<= darr[i] <=22):
                darr[i] = 75.
            else:
                darr[i] = 0.
        return darr
    if(n==3):
        for i in range(0,len(darr)):
            if(9 <= darr[i] <= 11) or (21 <= darr[i]  <= 23):
                darr[i] = 350.
            else:
                darr[i] = 0.
        return darr
    if(n==4):
        for i in range(0,len(darr)):
            darr[i] = 48.83 + 65
        return darr
    if(n==5):
        for i in range(0,len(darr)):
            if(15 <= darr[i] <= 15.5):
                darr[i] = 850.
            else:
                darr[i] = 0.
        return darr
    if(n==6):
        for i in range(0,len(darr)):
            if(8 <= darr[i] <= 12) or (15 <= darr[i] <= 20):
                darr[i] = 50
            else:
                darr[i] = 0
        return darr
    if(n==7):
        for i in range(0,len(darr)):
            if(0 <= darr[i] <= 6) or (22 <= darr[i] <= 24):
                darr[i] = 150
            if(8 <= darr[i] <= 11) or (12 <= darr[i] <= 17):
                darr[i] = 50
            else:
                darr[i] = 0
        return darr
    if(n==8):
        for i in range(0,len(darr)):
            if(19 <= darr[i] <= 22):
                darr[i] = 15
            else:
                darr[i] = 0
        return darr
    if(n==9):
        for i in range(0,len(darr)):
            if(8 <= darr[i] <= 10) or (12 <= darr[i]  <= 14) or (16 <= darr[i]  <= 18) or (20<= darr[i] <=22):
                darr[i] = 75.
            else:
                darr[i] = 0.
        return darr
    

def digitalplotter():
    #To plot the individual device plots
    plt.xlim(0,24)
    plt.ylim(0,1200)
    plt.grid()
    plt.xlabel("Time in Hours")
    plt.ylabel("Audited Power in Watts")
    plt.locator_params(axis="x", nbins=24)
    plt.locator_params(axis="y", nbins=24) 
    col = ['red','green','blue','orange','black','purple','magenta','cyan']    
    for i in range(1,9):
        a = digital(i,timeline)
        plt.plot(timeline,a,color = col[i-1])
    plt.legend(['1','2,3','4','5,10,11,12,13','6','7','8','9'], loc ="best")
    plt.show()

def sumplotter():
    #To plot the entire load curve adding all the individual device power consumption
    plt.xlim(0,24)
    plt.ylim(0,1200)
    plt.grid()
    plt.xlabel("Time in Hours")
    plt.ylabel("Audited Power in Watts")
    plt.locator_params(axis="x", nbins=24)
    plt.locator_params(axis="y", nbins=24) 
    c = digital(1,timeline) 
    for i in range(2,10):
        c = add(c,digital(i,timeline))  
    plt.plot(timeline,c,color='red')
    plt.show()
```

## Calling the functions to display output

```python
digitalplotter()
sumplotter()
```
